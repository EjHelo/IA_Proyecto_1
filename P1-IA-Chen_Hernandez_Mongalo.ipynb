```python
import pandas as pd
import numpy as np

dataframe = pd.read_csv("wisc_bc_data.csv",parse_dates=True)

dataframe.head(10)
```

## NormalizaciÃ³n

```python
dummies_diagnosis = pd.get_dummies(dataframe.diagnosis)
dummies_diagnosis.head(20)
```

```python
dataframe_temporal = pd.concat([dataframe, dummies_diagnosis], axis='columns')
dataframe_temporal.head(10)
```

```python
dataframe_normalizado = dataframe_temporal.drop(['diagnosis'], axis='columns')
dataframe_normalizado.head(10)
```

```python
columns = ['radius_mean','texture_mean','perimeter_mean','area_mean', 'smoothness_mean']

for var in range (len(columns)):
    old_var = columns[var]
    new_var = 'z_'+ old_var
    dataframe_normalizado[new_var] = (dataframe[old_var] - dataframe[old_var].mean())/dataframe[old_var].std(ddof=0)
dataframe_normalizado.head(10)
```
